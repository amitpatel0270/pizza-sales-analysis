```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
order_details = pd.read_csv(r"C:\Users\asus\Desktop\major_project\pizza_sales\pizza_sales\order_details.csv")
orders = pd.read_csv(r"C:\Users\asus\Desktop\major_project\pizza_sales\pizza_sales\orders.csv")
pizza_types = pd.read_csv(r"C:\Users\asus\Desktop\major_project\pizza_sales\pizza_sales\pizza_types.csv",encoding='latin1')
pizzas = pd.read_csv(r"C:\Users\asus\Desktop\major_project\pizza_sales\pizza_sales\pizzas.csv")
```

## Merging data.......

```python
df = order_details.merge(orders,on='order_id')
df = df.merge(pizzas,on = 'pizza_id')
df = df.merge(pizza_types,on='pizza_type_id')
df.head()
```

```python
df.info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['time'] = pd.to_datetime(df['time'],format = "%H:%M:%S").dt.time
```

```python
df.info()
```

## Create Matrices..... 



```python
df['revenue']= df['price']*df['quantity']

df['month'] = df['date'].dt.month_name()

df['weakday'] = df['date'].dt.day_name()

df['hour'] = pd.to_datetime(df['time'], format='%H:%M:%S', errors='coerce').dt.hour
```

```python
df.head()
```

## Exploratory Data Analysis

```python
Total_Revenue = df['revenue'].sum()
```

## Revenue by Category

```python
Category_sales = df.groupby('category')[['revenue']].sum().sort_values(by = 'revenue',ascending=False).reset_index()

Category_sales 
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(x= "category", y = "revenue" , data = Category_sales)
ax.bar_label(ax.containers[0])
plt.title('Revenue By Caregory')
plt.savefig("revenueBycategory")
plt.show()
```

## Top 10 best selleing pizza by revenue....

```python
top_pizza = df.groupby('name')[['revenue']].sum().nlargest(10,columns='revenue').reset_index()

top_pizza
```

```python
plt.figure(figsize=(15,5))
qx=sns.barplot(x='name',y='revenue',data = top_pizza)
plt.title('Top 10 selling pizza by revenue')
qx.bar_label(qx.containers[0])
plt.xticks(rotation= 65)
plt.show()
```

## Top 10 worst selling pizza by revenue....

```python
worst_pizza = df.groupby('name')[['revenue']].sum().sort_values(by='revenue',ascending=True).reset_index()
worst_pizza.head(10)
```

```python
plt.figure(figsize=(15,5))
qx=sns.barplot(x='name',y='revenue',data = worst_pizza.head(10))
plt.title('Top 10 worst selling pizza by revenue')
qx.bar_label(qx.containers[0])
plt.xticks(rotation= 65)
plt.show()
```

## Top 10 selling pizza by quantity....

```python
quantity_sales = df.groupby('name')[['quantity']].sum().sort_values(by='quantity',ascending=False).reset_index()
quantity_sales.head(10)
```

```python
plt.figure(figsize=(15,5))
qx=sns.barplot(x='name',y='quantity',data = quantity_sales.head(10))
plt.title('Top 10 selling pizza by quantity')
qx.bar_label(qx.containers[0])
plt.xticks(rotation= 65)
plt.show()
```

## Top 10 worst selling pizza by quantity......

```python
lowquantity_sales = df.groupby('name')[['quantity']].sum().sort_values(by='quantity',ascending=True).reset_index()
lowquantity_sales.head(10)
```

```python
plt.figure(figsize=(15,5))
qx=sns.barplot(x='name',y='quantity',data = lowquantity_sales.head(10))
plt.title('Top 10 worst selling pizza by quantity')
qx.bar_label(qx.containers[0])
plt.xticks(rotation= 65)
plt.show()
```

## Revenue by size.....

```python
size_sale = df.groupby('size')[['revenue']].sum().sort_values(by = 'revenue',ascending=False).reset_index()

size_sale
```

```python
plt.figure(figsize=(8,5))
px = sns.barplot(x = 'size', y = 'revenue',data = size_sale)
px.bar_label(px.containers[0])
plt.title("Revenue By Pizzas Size")
plt.show()
```

## Monthly Treand

```python
monthly_sales = df.groupby(df['date'].dt.to_period('M'))['revenue'].sum().reset_index()
monthly_sales.plot(kind='line', marker='o', figsize=(10,5), title='Monthly Revenue Trend')
monthly_sales

```
